```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot
from time import sleep
%matplotlib notebook

from data_collector import RESULT_FN, DQ_FN

def plot(fig, ax):
    results = pd.read_csv(RESULT_FN)
    dq = pd.read_csv(DQ_FN)
    # Filter dq's
    for dq in dq['contestant'].unique():
        results = results[results['contestant'] != dq]
    ax.set_xscale('linear')
    ax.set_yscale('linear')
    sns.lineplot(data=results, x='n', y='time', hue='contestant', ax=ax)
```

```python
fig, ax = pyplot.subplots(figsize=(15,10))
pyplot.ion()

fig.show()
fig.canvas.draw()
while True:
    ax.clear()
    plot(fig, ax)
    fig.canvas.draw()
#     sleep(2)
```
